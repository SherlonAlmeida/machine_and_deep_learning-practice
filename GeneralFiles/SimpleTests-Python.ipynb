#Testes para Q-Learning com Numpy

```python
import numpy as np
```

```python
table = np.zeros([5, 4])
```

```python
table
```

```python
table[0,1] = 5
table[2,2] = 27
table
```

```python
table[2, :] #Mostra o conteudo da linha 2
```

```python
np.argmax(table[2, :]) #Obtém o Index da coluna com o maior número
```

```python
acoes = ["Cima", "Baixo", "Esquerda", "Direita"]

#Ou seja, considerando que o melhor resultado foi o número 27, a resposta desejada é o Índice 2 que representa a ação 2 para o aprendizado por reforço.
#Se as ações fossem [cima, baixo, esquerda, direita], o índice 2 representaria "esquerda" para o agente.

print(f"A melhor ação a partir do estado atual é: {acoes[np.argmax(table[2, :])]}")
```

#Exportar e Carregar Numpy Array

```python
#Digamos que aprendemos a Q-table durante o treinamento e queremos salvar para o teste, podemos usar np.save()
np.save("table_learned.npy", table)
```

```python
#Agora que salvamos, podemos carregar com np.load()
table_loaded = np.load("table_learned.npy")
print(table_loaded)
```

#Removendo dados Nulos

```python
#Referência: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
import pandas as pd
data = [["Sherlon", 28],
        ["Maria", None],
        ["Naldo", 25],
        ["João", None],
        ["Lucas", 0]]
x = pd.DataFrame(data)
x.columns = ["Nome", "Idade"]
x
```

```python
#Remove os dados Nulos
x_filtered = x.dropna()
```

```python
x_filtered
```

#Removendo dados inconsistestes (Idade < 18)

```python
#Apresenta os dados que satisfazem a condição desejada, neste caso Idade >= 18.
x_filtered = x_filtered[x_filtered["Idade"] >= 18]
```

```python
x_filtered
```

#Exportar e Carregar o DataFrame filtrado

```python
import pandas as pd
import numpy as np

#Exportar os dados
saida = np.array(x_filtered)
np.save("meus_dados_filtrados.npy", saida)
```

```python
#Carregar os dados
entrada = np.load("meus_dados_filtrados.npy", allow_pickle=True)
x_carregado = pd.DataFrame(entrada)
x_carregado
```
